## Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt 
```

## Loading Dataset

```python
data =pd.read_csv("Iris.csv")
data.head()
```

```python
data.info()
```

```python
# removing the column Id because there is no need of it 
data = data.drop(['Id'],axis = 1)
```

```python
data.describe()
```

```python
X = data.iloc[:,:-1].values
y = data.iloc[:,-1].values
```

## Encoding Species Labels

```python
from sklearn.preprocessing import LabelEncoder
lab = LabelEncoder()
y=lab.fit_transform(y)
y
```

#### Iris-setosa = 0 <br> Iris-versicolor = 1 <br> Iris-virginica = 2

### Splitting training and testing data

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X,y)
```

## Building Model

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf = clf.fit(train_X,train_y)
print('Decision Tree Classifer Created')
```

```python
from sklearn import tree
fn=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn=['setosa','versicolor','virginica']

fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (4,4), dpi = 300)

tree.plot_tree(clf, feature_names = fn, class_names = cn, filled = True);
```

### Making Predictions

```python
pred = clf.predict(test_X)
pred
```

### Accuracy Check

```python
score = clf.score(test_X, pred)
score
```

```python
# Comparing Actual vs Predicted Values
df = pd.DataFrame({'Actual': test_y, 'Predicted': pred})  
df.head()
```
